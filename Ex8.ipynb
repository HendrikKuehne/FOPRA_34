```python
import qiskit
```

Circuit to measure $<\sigma_z>$

```python
q0 = qiskit.QuantumRegister(1, name = 'q0')
c0 = qiskit.ClassicalRegister(1, name = 'c0')

circ_8z = qiskit.QuantumCircuit(q0,c0)

circ_8z.z(q0)
circ_8z.measure(q0,c0)
circ_8z.draw('mpl')
```

Circuit to measure $<\sigma_x>$

```python
circ_8x = qiskit.QuantumCircuit(q0,c0)
circ_8x.h(q0)
circ_8x.z(q0)
circ_8x.h(q0)
circ_8x.measure(q0,c0)
circ_8x.draw('mpl')
```

Circuit to measure $<\sigma_y>$

```python
circ_8y = qiskit.QuantumCircuit(q0,c0)
circ_8y.s(q0)
circ_8y.h(q0)
circ_8y.z(q0)
circ_8y.h(q0)
circ_8y.s(q0)
circ_8y.measure(q0,c0)
circ_8y.draw('mpl')
```
